Data from [World Bank Open Data](https://data.worldbank.org) about distribution of female employment by sector in the United States.

```python
import pandas as pd
indus = pd.read_csv('IND EMPL FE.csv', index_col=0, skiprows=4)
agr = pd.read_csv('AGR EMPL FE.csv', index_col=0, skiprows=4)
srv = pd.read_csv('SRV EMPL FE.csv', index_col=0, skiprows=4)
```

```python
us_agr = agr.loc['United States','1985':'2016']
us_indus = indus.loc['United States','1985':'2016']
us_srv = srv.loc['United States','1985':'2016']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
sns.set_style("white")

fig = plt.figure(figsize=(10, 7))
ax = plt.gca()
ax.plot(range(1985,2017), us_agr, '-', color= sns.color_palette("Greens")[2], label='Agriculture')
ax.plot(range(1985,2017), us_indus, '-', color= sns.color_palette("Reds")[2], label='Industry')
ax.plot(range(1985,2017), us_srv, '-', color= sns.color_palette("Blues")[2], label='Services')

ax.set_xlabel('Years', labelpad=12)
ax.set_ylabel('%', labelpad=20)
ax.set_title('Distribution of female employment by sector in the US')
plt.xticks(alpha=0.4)
plt.yticks(alpha=0.4)
sns.despine()
ax.legend()
#fig.savefig("temp.png", dpi=500)
```
